Extração dos dados

```python
import pandas
import pandera
```

```python
dataframe = pandas.read_csv("ocorrencia_2010_2020.csv", sep=";",parse_dates=["ocorrencia_dia","divulgacao_dia_publicacao"],dayfirst=True)
dataframe.head(10)
```

```python
dataframe.dtypes
```

```python
dataframe.ocorrencia_dia.dt.month
```

Validação da Extração dos dados

```python
schema = pandera.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pandera.Column(pandera.Int),
        "codigo_ocorrencia2":pandera.Column(pandera.Int),
        "ocorrencia_classificacao":pandera.Column(pandera.String),
        "ocorrencia_cidade":pandera.Column(pandera.String),
        "ocorrencia_uf":pandera.Column(pandera.String),
        "ocorrencia_aerodromo":pandera.Column(pandera.String),
        "ocorrencia_dia":pandera.Column(pandera.DateTime),
        "ocorrencia_hora":pandera.Column(pandera.String,pandera.Check.str_matches(r'([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'),nullable=True),
        "total_recomendacoes":pandera.Column(pandera.Int)
        
    }
)
```

```python
schema.validate(dataframe)
```

Limpeza dos dados

```python
dataframe.head()
```

```python
dataframe.replace(['***','###!','####','****','*****','NULL','NaN'],pandas.NA, inplace=True)
```

```python
dataframe.isnull().sum()
```

```python
dataframe.drop(['codigo_ocorrencia1','codigo_ocorrencia3','codigo_ocorrencia4','ocorrencia_latitude','ocorrencia_longitude','ocorrencia_pais','investigacao_aeronave_liberada','investigacao_status','divulgacao_relatorio_numero','divulgacao_relatorio_publicado','divulgacao_dia_publicacao','total_aeronaves_envolvidas','ocorrencia_saida_pista'],axis=1,inplace=True)
```

```python
dataframe.head()
```

```python
dataframe.loc[dataframe.ocorrencia_uf.isnull()]
```

```python
dataframe.dropna(subset=['ocorrencia_uf'],inplace=True)
```

```python
dataframe
```

```python
#Linhas em que o incidente é grave e ocorre em SP
incidente_grave = dataframe.ocorrencia_classificacao == 'INCIDENTE GRAVE'
ocorrencia_sp = dataframe.ocorrencia_uf == 'SP'
dataframe.loc[incidente_grave & ocorrencia_sp]
```

```python
#Linhas em que o incidente é grave ou ocorre em SP
incidente_grave = dataframe.ocorrencia_classificacao == 'INCIDENTE GRAVE'
ocorrencia_sp = dataframe.ocorrencia_uf == 'SP'
dataframe.loc[incidente_grave | ocorrencia_sp]
```

```python
#ocorrências cuja (classificação == INCIDENTE GRAVE ou classificação == INCIDENTE) e o estado == SP
incidente = dataframe.ocorrencia_classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])
ocorrencia_sp = dataframe.ocorrencia_uf == 'SP'
dataframe.loc[incidente & ocorrencia_sp]
```

```python
#ocorrencias cuja cidade comecem com a letra "c"
filtro = dataframe.ocorrencia_cidade.str.startswith('C')
dataframe.loc[filtro]
```

```python
#ocorrencias cuja cidade terminem com a "MA"
filtro = dataframe.ocorrencia_cidade.str.endswith('MA')
dataframe.loc[filtro]
```

```python
#ocorrencias cuja cidade contém os caracteres "MA"
filtro = dataframe.ocorrencia_cidade.str.contains('MA')
dataframe.loc[filtro]
```

```python
#ocorrências do ano de 2015
filtro = dataframe.ocorrencia_dia.dt.year == 2015
dataframe.loc[filtro]
```

```python
#ocorrências do ano de 2015 e mês 12
filtro = dataframe.ocorrencia_dia.dt.year == 2015
filtro2 = dataframe.ocorrencia_dia.dt.month == 12
dataframe.loc[filtro & filtro2]
```

```python
#ocorrências do ano de 2015 e mês 12 e dia 8
filtro_ano = dataframe.ocorrencia_dia.dt.year == 2015
filtro_mes = dataframe.ocorrencia_dia.dt.month == 12
filtro_dia = dataframe.ocorrencia_dia.dt.day == 8
dataframe.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
#ocorrências do ano de 2015 e mês 12 e entre o dia 5 e 8
filtro_ano = dataframe.ocorrencia_dia.dt.year == 2015
filtro_mes = dataframe.ocorrencia_dia.dt.month == 12
filtro_dia = (dataframe.ocorrencia_dia.dt.day >= 5) & (dataframe.ocorrencia_dia.dt.day <= 8)
dataframe.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
dataframe['ocorrencia_dia_hora'] = pandas.to_datetime(dataframe.ocorrencia_dia.astype(str) + ' ' + dataframe.ocorrencia_hora)
dataframe.head()
```
