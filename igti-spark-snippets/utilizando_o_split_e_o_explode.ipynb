```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as f
```

```python
spark = SparkSession.builder.appName('IGTI').getOrCreate()
```

```python
df = spark.read.parquet('imdb_parquet/')
```

```python
(
    df
    .limit(5)
    .toPandas()
)
```

```python
(
    df
    .replace(r'\N', None)
    .withColumn('genres_array', f.split(f.col('genres'),','))
    .withColumn('genres_unique', f.explode(f.col('genres_array')))
    .limit(20)
    .toPandas()
)
```
